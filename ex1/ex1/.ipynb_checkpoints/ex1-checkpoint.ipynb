# Unvariable Cost Function

```python
def cost_function(X,y,theta):
    
    #theta0,theta1
    temp=np.dot(X,theta)
    temp=temp-y
    m=X.shape[0]
    temp=temp**2
    temp=temp/2
    temp=temp/m
    return np.sum(temp)
    
```

```python
theta=np.zeros((2,))
J=cost_function(X,y,theta)
print(J),print(theta)
```

```python
def gradient_descent(X,y,theta):
    alpha=0.01
    iters=1500
    m=len(y)
        
    for i in range(1500):
        temp=np.dot(X,theta)
        temp=temp-y
        temp2 =np.dot(X.transpose(),temp)
        temp2 =temp2/m
        theta =theta- alpha*temp2
    return theta
```

```python
theta_f=gradient_descent(X,y,theta)
```

```python
theta_f
```

```python
theta2=np.zeros((2,))
theta2[0]=-1
theta2[1]=2

J=cost_function(X,y,theta2)
J
```
